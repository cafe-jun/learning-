```python
import matplotlib.pyplot as plt
```

```python
#1ch plot chart

plt.plot([1,2,3],[5,7,4])
plt.show()
```

```python
#2ch 2ray plot chart
x= [1,2,3]
y= [5,7,10]

x2= [1,2,3]
y2= [11,15,17]
plt.plot(x,y,label='first line')
plt.plot(x2,y2,label='second lline')

plt.xlabel('Plot Number')
plt.ylabel('import var')

plt.title('junseok\njajajja') #\n 줄 넘기기 
plt.legend()
plt.show()
```

```python
#3ch bar chart
x= [2,4,6,8,10]
y= [6,7,8,5,3]

x2= [1,3,5,7,9,11]
y2= [5,6,2,7,3,3]
plt.bar(x,y,label='bars1',color='r')
plt.bar(x2,y2,label='bars2',color='g')
plt.xlabel('x')
plt.ylabel('y')
plt.title('junseok\n 22a')
plt.legend()
plt.show()
```

```python
#3ch multi bar chart 
population_age=[22,55,63,22,35,60,55,35,46,234,102,10,15,32,33,45,67,12,65,31,111,53,123,35]
# y값 
ids = [x for x in range(len(population_age))] #x축 만들기 

#print(ids)
plt.bar(ids, population_age)
plt.xlabel('x')
plt.ylabel('y')
plt.title('junseok\n 22a')
plt.show()
```

```python
#4ch scatter plots 
x= [1,2,3,4,5,6,7]
y= [5,3,4,3,1,2,4]
plt.scatter(x,y,label='skitscat',color='k',s=50,marker='x')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#5ch stack plots
days = [1,2,3,4,5]

sleeping = [7,8,9,10,11]
eating = [2,3,4,5,3]
working = [7,8,7,2,2]
playing = [8,5,4,2,13]

plt.plot([],[],color='m',label='sleeping')
plt.plot([],[],color='c',label='eating')
plt.plot([],[],color='k',label='working')
plt.plot([],[],color='b',label='playing')


plt.stackplot(days,sleeping,eating,working,playing ,colors=['m','c','k','b'])


plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
#6ch Pie charts



sleeping = [7,8,9,10,11]
eating = [2,3,4,5,3]
working = [7,8,7,2,2]
playing = [8,5,4,2,13]

slices = [7,2,2,13]
activate=['sleeping','eating','working','playing']
cols = ['c','m','r','b']
plt.pie(slices, labels=activate, 
        colors=cols,
        startangle=90,
        shadow=True,
        explode=(0,1,0,0),
         autopct='%1.1f%%')

#plt.xlabel('x')
#plt.ylabel('y')
#plt.legend()
plt.show()
```

```python
#7ch loading file 
#part 1 
import csv 

with open ('example.txt','r') as csvfile:
    plots = csv.reader(csvfile,delimiter=',')
    for row in plots :
        x.append(int(row[0]))
        y.append(int(row[1]))

plt.plot(x,y,label="loading file ")
plt.show()
```

```python
#part 2
import numpy as np

x,y,=np.loadtxt('example.txt',delimiter=',',unpack=True)
plt.plot(x,y,label="loading file ")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
